# Polynomial Regression

**Importing Necessary Files**

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.preprocessing
import seaborn as sns
%matplotlib inline
sns.set()
```

# Import the 'Boston.csv' data file

```python
# write code here
data = pd.read_csv('Boston.csv')
data.head()
```

```python
# Write code here
df = data.copy()
```

```python
# Write code here
df.head()
```

### Creating Independent and Dependent Variables

```python
# Write code here
X = df.drop(['median home price'], axis=1)
Y = df[['median home price']]
print(X.shape)
print(Y.shape)
```

### Normalizing Data

```python
# Write code here
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
x_scale_minmax=min_max_scaler.fit_transform(X)
```

```python
X=pd.DataFrame(data=x_scale_minmax, columns=['Crime Rate','Residential Proportion','non-retail business acres/Town','Charles River',
                                             'NO2 concentration','Average Rooms/Dwelling.','Prior Built Units Proportion','Distance to Employment Centres','Radial Highways Distance','ValueProperty/tax rate','Teacher/town','blacks/town','Lower Status Percent'
])
X.head()
```

```python
X.describe()
```

### Splitting Data

```python
# Write code here
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.30, random_state=25,shuffle=True)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### Applying Linear Regression

```python
#Write the code here
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Apply the fit() function on the model

```python
#Write code here
model.fit(x_train,y_train)
```

Get the predicted values on training data set in train_pred

```python
#Write code here
train_pred = model.predict(x_train)
train_pred
```

Get the predicted values on test data set in test_pred

```python
#Write code here
test_pred= model.predict(x_test)
test_pred
```

### Finding R2 Score

```python
#Write code here for testing data
from sklearn.metrics import r2_score
r2_test_lr=r2_score(y_test,test_pred)
r2_test_lr
```

```python
#Write code here for training data
r2_train_lr=r2_score(y_train,train_pred)
r2_train_lr
```

# Applying Polynomial Regression 

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features=PolynomialFeatures(degree=2)
```

```python
# transform the features to higher degree features.
x_train_poly=poly_features.fit_transform(x_train)
```

```python
x_train_poly.shape
```

```python
# fit the transformed features to Linear Regression
poly_model=LinearRegression()
poly_model.fit(x_train_poly,y_train)
```

```python
# predicting on training data-set
y_train_predicted=poly_model.predict(x_train_poly)
```

```python
# predicting on test data-set
x_test_poly=poly_features.fit_transform(x_test)
y_test_predict=poly_model.predict(x_test_poly)
```

```python
# evaluating the model on training dataset
r2_train = r2_score(y_train, y_train_predicted)
```

```python
# evaluating the model on test dataset
r2_test = r2_score(y_test, y_test_predict)
```

```python
print ('The r2 score for training set is: ',r2_train)
print ('The r2 score for testing set is: ',r2_test)
```

# Model Complexity vs Performance

```python
from sklearn.metrics import mean_squared_error,make_scorer
from sklearn.model_selection import cross_val_score
mse_train=[]
mse_test=[]
mse_cv=[]
r2_train=[]
r2_test=[]
r2_cv=[]
for i in range(1,8):
    poly_reg = PolynomialFeatures(degree=i)
    
    X_tr_poly,X_tst_poly= poly_reg.fit_transform(x_train),poly_reg.fit_transform(x_test)
    
    poly = LinearRegression()
    poly.fit(X_tr_poly, y_train)
   
    y_tr_predicted,y_tst_predict = poly.predict(X_tr_poly),poly.predict(X_tst_poly)
    
    mse_cv_scores=cross_val_score(poly,X_tr_poly,y_train,cv=5,scoring=make_scorer(mean_squared_error))
    r2_cv_scores=cross_val_score(poly,X_tr_poly,y_train,cv=5)
    
    r2_train.append(r2_score(y_train, y_tr_predicted))
    r2_test.append(r2_score(y_test, y_tst_predict))
    r2_cv.append(r2_cv_scores.mean())
    
    mse_train.append(mean_squared_error(y_train, y_tr_predicted))
    mse_test.append(mean_squared_error(y_test, y_tst_predict))
    mse_cv.append(mse_cv_scores.mean())
```

### Comparison of R2 Score

```python
plt.figure(figsize=(18,5))
sns.lineplot(x=list(range(1,8)), y=r2_train, label='Training')
sns.lineplot(x=list(range(1,8)), y=r2_test, label='Testing')
sns.lineplot(x=list(range(1,8)), y=r2_cv, label='Cross Validation')
plt.ylim(0,1.1)
plt.show()
```

### Comparison of Mean Squared Error

```python
plt.figure(figsize=(15,4))
sns.lineplot(x=list(range(1,8)), y=mse_train, label='Training')
sns.lineplot(x=list(range(1,8)), y=mse_test, label='Testing')
sns.lineplot(x=list(range(1,8)), y=mse_cv, label='Cross Validation')
plt.ylim(0,70)
plt.show();
```
